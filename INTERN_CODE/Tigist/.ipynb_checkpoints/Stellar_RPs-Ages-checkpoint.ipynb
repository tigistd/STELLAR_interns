```python
%matplotlib inline
import numpy as np
import pandas as pd
from astropy.io import ascii
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
keplerdata = pd.read_csv('Table_1_Periodic.txt')
keplerdata
```

```python
table2 = pd.read_csv('Table_2_Non_Periodic.txt')
table2
```

```python
ourdata = pd.read_csv('measured_rps.csv')
#ourdata
```

```python
danielles = Table.read('Interns_tic_table.fits')
danielles.colnames
```

```python
count = 0
for row in danielles:
    #print(row)
    #break
    tic=row['ID']
    ourtics=ourdata['tic'].values
    if tic in ourtics:
        print(count)
        count+=1
    else:
        pass
```

```python
#pulling values from danielles table to our table
#empty lists to hold values
our_tics=[]
our_temps=[]
ourgaiabp=[]
ourgaiarp=[]
##for loop to find/store
#count=0
for row in danielles:
    tic=row['ID']
    temp=row['Teff']
    bp=row['gaiabp']
    rp=row['gaiarp']
    #define our data
    ourtics=ourdata['tic'].values
    #check if danielles's data's row matches
    if tic in ourtics:
        #print(count)
        #count +=1
        our_tics.append(tic)
        our_temps.append(temp)
        ourgaiabp.append(bp)
        ourgaiarp.append(rp)
    else:
        pass
```

```python
data_to_add= {'tic': our_tics, 'temp': our_temps,'bp': ourgaiabp,'rp':ourgaiarp}
df_to_add=pd.DataFrame(data=data_to_add)
df_to_add
```

```python
#use pd.merge, parameters 'sort'& 'on'
ourdata_wtemp=pd.merge(ourdata, df_to_add,sort=True,on='tic')
```

```python
ourdata_wtemp
```

```python
#graph w rp as y and temp as x
plt.figure(figsize=(10,5),dpi=200)
plt.scatter(keplerdata['Teff'],keplerdata['Prot'],s=.1,c='b',label = 'Table1')
plt.scatter(ourdata_wtemp['temp'],ourdata_wtemp['period'],s=20,c='r',label='our data')
plt.xlim(6500,3500)
plt.xlabel("Temperature",fontsize=20)
plt.ylabel('Rotation Period',fontsize=20)
plt.title('Stellar Rotation Periods', fontsize=25)
plt.legend(loc='upper right')
```
