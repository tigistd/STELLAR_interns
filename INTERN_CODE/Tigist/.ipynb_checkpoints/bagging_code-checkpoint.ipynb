```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
from sklearn.tree import DecisionTreeClassifier
import random
from sklearn import tree
```

```python
#generate fake data
#create 5 empty lists (4 for features and 1 for labels)
#start a for loop that cycles through the range (0, 100)
#generate ranfom data to an empty list
#end for loop
#print the final lengths
```

```python
#features
temps=[]
spec_type=[]
luminosity=[]
mass=[]
#label
ages=[]
for i in range(0,100):
    temp_ran=random.randint(3500, 60000)
    spec_ran=random.randint(0,7)
    lum_ran=random.randint(1, 1000000)
    mass_ran=random.randint(1, 100)
    ages_ran=random.randint(100000, 13000000)
    finaltemp=temps.append(temp_ran)
    finalspec=spec_type.append(spec_ran)
    finallum=luminosity.append(lum_ran)
    finalmass=mass.append(mass_ran)
    finalages=ages.append(ages_ran)
print(len(temps), len(spec_type),len(luminosity),len(mass),len(ages))
```

```python
data={'temps':temps, 'spec_type':spec_type,
                  'luminosity':luminosity, 'mass': mass}
```

```python
df = pd.DataFrame(data=data)
df
```

```python
data_table = Table([temps,spec_type,luminosity,mass],names=('temps','spec_type','luminosity','mass'))
```

```python
random.seed(2)
indices=[]
for i in range(0,10):
    indices.append(random.randint(0,100))
data_subset=[]
labels_subset=[]
for i in indices:
    print(i)
    data_subset.append(data_table[i])
    labels_subset.append(data_table[i])
```

```python
random.seed(5)
indices=[]
for i in range(0,10):
    indices.append(random.randint(0,100))
data_subset=[]
labels_subset=[]
for i in indices:
    print(i)
    data_subset.append(data_table[i])
    labels_subset.append(data_table[i])
```

```python
model = DecisionTreeClassifier()
subset_tree_fit=model.fit(data_subset,labels_subset)
```

```python
plt.figure(figsize=(20,15))
tree.plot_tree(subset_tree_fit)
```
