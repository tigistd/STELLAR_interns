```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("sunspots_data2.csv")
data
```

```python
years = data["year"]
daily_sun = data["daily_sunspots"]
```

```python
import matplotlib.pyplot as plt
plt.bar(years,daily_sun)#, width = 1)
plt.title("daily sunspots over time")
plt.ylabel("sunspots")
plt.show()
```
